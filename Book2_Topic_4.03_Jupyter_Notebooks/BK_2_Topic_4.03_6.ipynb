# 堆叠面积图
Book_2《可视之美》 | 鸢尾花书：从加减乘除到机器学习  

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 如果文件夹不存在，创建文件夹
if not os.path.isdir("Figures"):
    os.makedirs("Figures")
```

```python
df = pd.read_excel('populations_over_continents.xlsx', index_col=0)
# 读入Excel表格
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax = df.plot.area(xlim = (1950, 2015), alpha = 0.4, ax = ax, ylabel = 'Million')

fig.savefig('Figures/堆叠面积图.svg', format='svg')
```

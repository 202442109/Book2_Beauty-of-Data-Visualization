# 使用patches绘制平面几何元素
Book_2《可视之美》 | 鸢尾花书：从加减乘除到机器学习  

```python
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon, Circle, CirclePolygon
import numpy as np

import os

# 如果文件夹不存在，创建文件夹
if not os.path.isdir("Figures"):
    os.makedirs("Figures")
```

```python
fig, axs = plt.subplots(ncols = 1, nrows = 4)

for num_vertices, ax in zip([4,5,6,8], axs.ravel()):
    
    hexagon_inner = RegularPolygon((0,0), numVertices=num_vertices, 
                                   radius=1, alpha=0.2, edgecolor='k')
    ax.add_patch(hexagon_inner)
    # 绘制正圆内接多边形
    
    hexagon_outer = RegularPolygon((0,0), numVertices=num_vertices, 
                                   radius=1/np.cos(np.pi/num_vertices), 
                                   alpha=0.2, edgecolor='k')
    ax.add_patch(hexagon_outer)
    # 绘制正圆外切多边形
    
    circle = Circle((0,0), radius=1, facecolor = 'none', edgecolor='k')
    ax.add_patch(circle)
    # 绘制正圆
    
    ax.set_xlim(-1.5,1.5)
    ax.set_ylim(-1.5,1.5)
    
    ax.set_aspect('equal', adjustable='box')
    ax.axis('off')
    
fig.savefig('Figures/使用patches绘制平面几何元素.svg', format='svg')
```

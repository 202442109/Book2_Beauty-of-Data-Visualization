### 自定义函数

```python
def comb(n, k):
    # 产生系数
    return factorial(n) // (factorial(k) * factorial(n-k))

def get_bezier_curve(points):
    n = len(points) - 1
    # 贝塞尔曲线阶数
    # 匿名函数，产生特定比例 t 对应的贝塞尔曲线坐标
    # 下式相当于加权平均
    return lambda t: sum(comb(n, i)*t**i * (1-t)**(n-i)*points[i] for i in range(n+1))

def evaluate_bezier(points, total):
    # 产生连续贝塞尔曲线坐标点
    bezier = get_bezier_curve(points)
    new_points = np.array([bezier(t) for t in np.linspace(0, 1, total)])
    return new_points[:, 0], new_points[:, 1]
```

### 二阶

```python
A = [0, 0]
B = [1, 1]
# 曲线的两个端点

fig = plt.figure(figsize = (10,16), tight_layout=True)

for idx in range(18):
    
    P1 = [rnd.random(), rnd.random()]
    # 用随机数生成控制点P1

    points = np.array([A, 
                       P1, 
                       B])
    
    ax = fig.add_subplot(6, 3, idx + 1)
    x, y = points[:, 0], points[:, 1]
    bx, by = evaluate_bezier(points, 200)
    # 
    plt.plot(bx, by, 'b-')
    plt.plot(x, y, 'r.')
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_aspect('equal', adjustable='box')

fig.savefig('Figures/二阶.svg', format='svg')
```

### 三阶

```python


fig = plt.figure(figsize = (10,16), tight_layout=True)

for idx in range(18):
    
    P1 = [rnd.random(), rnd.random()]
    P2 = [rnd.random(), rnd.random()]
    # 随机数生成两个控制点：P1、P2

    points = np.array([A, 
                       P1, 
                       P2, 
                       B])
    
    ax = fig.add_subplot(6, 3, idx + 1)
    x, y = points[:, 0], points[:, 1]
    bx, by = evaluate_bezier(points, 200)
    # 
    plt.plot(bx, by, 'b-')
    plt.plot(x, y, 'r.')
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_aspect('equal', adjustable='box')

fig.savefig('Figures/三阶.svg', format='svg')
```

### 四阶

```python
fig = plt.figure(figsize = (10,16), tight_layout=True)

for idx in range(18):
    
    P1 = [rnd.random(), rnd.random()]
    P2 = [rnd.random(), rnd.random()]
    P3 = [rnd.random(), rnd.random()]

    points = np.array([A, 
                       P1, 
                       P2, 
                       P3,
                       B])
    
    ax = fig.add_subplot(6, 3, idx + 1)
    x, y = points[:, 0], points[:, 1]
    bx, by = evaluate_bezier(points, 200)
    # 
    plt.plot(bx, by, 'b-')
    plt.plot(x, y, 'r.')
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_aspect('equal', adjustable='box')

fig.savefig('Figures/四阶.svg', format='svg')
```
